```python
import matplotlib.pyplot as plt
ours_color = 'blue'
default_color = 'red'
baseline_color = 'green'
```

```python
import matplotlib.pyplot as plt

# Data for plotting
threads = ["2", "4", "6", "8"]
default_latencies = [258, 193, 128, 789]
ours_latencies = [268, 208, 140, 567]

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(threads, default_latencies, marker='o', label='Default', color=default_color)
plt.plot(threads, ours_latencies, marker='o', label='Ours', color=ours_color)
plt.title("Wakeup Latencies 99.9th Percentile (usec)")
plt.xlabel("Number of Threads")
plt.ylabel("Latency (us)")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("wakeup_latencies.pdf")
plt.savefig("wakeup_latencies.png")
```

```python
# Data for 99.5% latencies
workers = ["2", "4", "6", "8"]
default_latencies = [1696.532, 874.962, 1176.855, 11577.021]
ours_latencies = [1699.312, 874.485, 622.259, 630.864]

# Plotting the line chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(workers, default_latencies, marker='o', label='Default', color=default_color)
plt.plot(workers, ours_latencies, marker='o', label='Ours', color=ours_color)
plt.title("99.5th Percentile Latencies (us)")
plt.xlabel("Number of Workers")
plt.ylabel("Latency (us)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.savefig("99.5_percentile_latencies.pdf")
plt.savefig("99.5_percentile_latencies.png")
```

```python
# Data for Target Throughput and Actual Throughput
workers = ["2", "4", "6", "8"]
target_throughput = [598748.615, 1197497.231, 1796245.846, 2394994.462]
default_actual = [407045, 762581.333, 814496.667, 319884.333]
ours_actual = [420958.333, 783750, 953805.667, 1430831.333]

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(workers, target_throughput, marker='o', linestyle='--', label='Target Throughput', color=baseline_color)
plt.plot(workers, default_actual, marker='o', label='Default Actual Throughput', color=default_color)
plt.plot(workers, ours_actual, marker='o', label='Ours Actual Throughput', color=ours_color)

plt.title("Target vs Actual Throughput")
plt.xlabel("Number of Workers")
plt.ylabel("Throughput")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the chart
plt.savefig("target_vs_actual_throughput.pdf")
plt.savefig("target_vs_actual_throughput.png")
```
